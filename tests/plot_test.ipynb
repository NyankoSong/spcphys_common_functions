```python
import spcphys_common_functions as scf
import numpy as np
from matplotlib import pyplot as plt
import time
```

```python
fig, ax = plt.subplots()

quadmesh = scf.plot_tools.plot_hist2d(ax, np.random.randn(1000), np.random.randn(1000), np.random.randn(1000), color_norm_range=(-3, 3), hist_pcolormesh_kwargs={'cmap':'RdBu'})
fig.colorbar(quadmesh)
```

```python
fig, ax = plt.subplots()

quadmesh = scf.plot_tools.plot_hist2d(ax, np.random.randn(1000), np.random.randn(1000), hist_pcolormesh_kwargs={'cmap':'RdBu_r'})
fig.colorbar(quadmesh)
```

```python
n_points = int(1E7)
x = np.sort(np.random.randn(n_points))
y = np.sin(x*2*np.pi) + np.random.randn(n_points)*0.1
```

```python
fig, ax = plt.subplots(figsize=(20, 3))
time_0 = time.time()
ax.plot(x, y, marker='.')
time_1 = time.time()

print('Time for normal plot: ', time_1 - time_0)
```

```python
fig, ax = plt.subplots(figsize=(20, 3))
time_1 = time.time()
scf.plot_tools.plot(x, y, axes=ax, max_dens=20, marker='x', ls='')
time_2 = time.time()

print('Time for scf.plot_tools.plot: ', time_2 - time_1) # Although this time is longer than ax.plot, the time to show the figure is much shorter, resulting in a faster user experience
```
